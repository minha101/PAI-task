##  Implement the Water Jug Problem using Depth-First Search
(DFS) and print the rules applied at each step.

```python
def water_jug_dfs():
    # Problem setup
    max1, max2, target = 4, 3, 2
    stack = [((0, 0), [])]  # (state, rules_list)
    visited = []
    
    print("Water Jug Problem - DFS Solution")
    print(f"Goal: Get {target} liters using {max1}L and {max2}L jugs\n")
    
    while stack:
        # Get current state (pop from end for DFS)
        (jug1, jug2), rules = stack.pop()
        
        # Skip if visited
        if (jug1, jug2) in visited:
            continue
        visited.append((jug1, jug2))
        
        print(f"State: ({jug1}, {jug2})")
        
        # Check goal
        if jug1 == target or jug2 == target:
        
            print("SOLUTION FOUND!")
            for i, r in enumerate(rules, 1):
                print(f"{i}. {r}")
            return True
        
        # Generate new states
        next_moves = []
        
        # Fill jug1
        if jug1 < max1:
            next_moves.append(((max1, jug2), f"Fill jug1: ({jug1},{jug2})-({max1},{jug2})"))
        
        # Fill jug2
        if jug2 < max2:
            next_moves.append(((jug1, max2), f"Fill jug2: ({jug1},{jug2})-({jug1},{max2})"))
        
        # Empty jug1
        if jug1 > 0:
            next_moves.append(((0, jug2), f"Empty jug1: ({jug1},{jug2})-(0,{jug2})"))
        
        # Empty jug2
        if jug2 > 0:
            next_moves.append(((jug1, 0), f"Empty jug2: ({jug1},{jug2})-({jug1},0)"))
        
        # Pour jug1 to jug2
        if jug1 > 0 and jug2 < max2:
            pour = min(jug1, max2 - jug2)
            next_moves.append(((jug1-pour, jug2+pour), f"Pour jug1->jug2: ({jug1},{jug2})->({jug1-pour},{jug2+pour})"))
        
        # Pour jug2 to jug1
        if jug2 > 0 and jug1 < max1:
            pour = min(jug2, max1 - jug1)
            next_moves.append(((jug1+pour, jug2-pour), f"Pour jug2->jug1: ({jug1},{jug2})->({jug1+pour},{jug2-pour})"))
        
        # Add to stack (reverse order for better exploration)
        for state, rule in reversed(next_moves):
            if state not in visited:
                stack.append((state, rules + [rule]))
    
    print("No solution exists")
    return False

# Run
water_jug_dfs()
```
