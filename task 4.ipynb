## Implement the N-Queens Problem (Dynamic)

```python
class NQueensSolver:
    def __init__(self):
        self.n = 0
        self.board = []
        self.solutions = []
        
    def solve(self):
        ##Main method to solve N-Queens problem

        print("N-QUEENS PROBLEM SOLVER")
    
        
        # Take input from user
        self.n = int(input("Enter number of queens (N): "))
        
        if self.n <= 0:
            print("Please enter a positive number!")
            return
        
        # Initialize empty board
        self.board = [-1] * self.n
        self.solutions = []
        
        print(f"\nSolving {self.n}-Queens problem...")
        print("(Each queen is placed in a separate row)")
        
        # Start solving from first row
        self.place_queen(0)
        
        # Display results
        self.display_results()
    
    def is_safe(self, row, col):
        #Check if queen can be placed at given position
        # Check all previous rows
        for prev_row in range(row):
            prev_col = self.board[prev_row]
            
            # Check same column
            if prev_col == col:
                return False
            
            # Check diagonal
            if abs(prev_row - row) == abs(prev_col - col):
                return False
        
        return True
    
    def place_queen(self, row):
        #Recursive function to place queens
        # If all queens placed, we found a solution
        if row == self.n:
            # Make a copy of current board
            solution = self.board.copy()
            self.solutions.append(solution)
            print(f"Found Solution #{len(self.solutions)}: {solution}")
            return True
        
        # Try placing queen in each column of current row
        found_solution = False
        for col in range(self.n):
            if self.is_safe(row, col):
                # Place queen
                self.board[row] = col
                
                # Recur for next row
                self.place_queen(row + 1)
                
                # Backtrack (remove queen)
                self.board[row] = -1
        
        return found_solution
    
    def display_results(self):
        #Display all solutions found
        print(f"RESULTS FOR {self.n}-QUEENS PROBLEM")
        
        if not self.solutions:
            print(f"No solutions exist for {self.n} queens!")
            return
        
        print(f"Total Solutions Found: {len(self.solutions)}")
    
        
        # Show first few solutions (max 5)
        show_count = min(5, len(self.solutions))
        for i in range(show_count):
            self.print_board(self.solutions[i], i + 1)
        
        if len(self.solutions) > 5:
            print(f"\n... and {len(self.solutions) - 5} more solutions")
    
    def print_board(self, solution, solution_num):
        #Print a single solution as a chessboard
        print(f"\nSolution #{solution_num}:")
        print("   " + " ".join([str(i+1) for i in range(self.n)]))
        print("   " + "-" * (2 * self.n))
        
        for row in range(self.n):
            queen_col = solution[row]
            print(f"{row+1} |", end=" ")
            
            for col in range(self.n):
                if col == queen_col:
                    print("Q", end=" ")
                else:
                    print(".", end=" ")
            print("|")
        
        print("   " + "-" * (2 * self.n))

# Simple version without class
def simple_n_queens():
#Simpler implementation of N-Queens
    
    # Get input
    n = int(input("How many queens? "))
    
    if n <= 0:
        print("Invalid input!")
        return
    
    # Arrays to store queen positions
    queens = [-1] * n  # queens[row] = column
    solutions = []
    
    def can_place(row, col):
        #Check if position is safe
        for r in range(row):
            c = queens[r]
            # Same column or diagonal
            if c == col or abs(r - row) == abs(c - col):
                return False
        return True
    
    def solve_from(row):
        #Try to place queens from given row
        if row == n:
            # Found a solution
            solutions.append(queens.copy())
            print(f"Solution {len(solutions)}: {queens}")
            return
        
        # Try each column
        for col in range(n):
            if can_place(row, col):
                queens[row] = col
                solve_from(row + 1)
                # Backtrack
                queens[row] = -1
    
    print(f"\nSolving {n}-Queens...")
    print("Rows are numbered 1 to N, Columns are numbered 1 to N")
    print("-" * 40)
    
    # Start solving
    solve_from(0)
    
    # Show results
    print("\n" + "=" * 40)
    print(f"Total Solutions: {len(solutions)}")
    
    if solutions:
        print("\nFirst solution (visual):")
        show_solution(solutions[0], n)

def show_solution(queens, n):
    #Display solution as chessboard
    print("\n   " + " ".join([str(i+1) for i in range(n)]))
    for row in range(n):
        print(f"{row+1} |", end=" ")
        for col in range(n):
            if col == queens[row]:
                print("Q", end=" ")
            else:
                print(".", end=" ")
        print("|")

# Another different approach using list of lists
def n_queens_alternative():
    #Alternative implementation using 2D board
    
    print("N-QUEENS PROBLEM")
    
    # Get N from user
    N = int(input("Enter value of N: "))
    
    if N < 1:
        print("N must be positive!")
        return
    
    # Create empty board
    board = []
    for i in range(N):
        row = []
        for j in range(N):
            row.append(0)
        board.append(row)
    
    solutions = []
    
    def is_position_safe(board, row, col):
        #Check if position is safe
        # Check same column above
        for i in range(row):
            if board[i][col] == 1:
                return False
        
        # Check upper left diagonal
        i, j = row, col
        while i >= 0 and j >= 0:
            if board[i][j] == 1:
                return False
            i -= 1
            j -= 1
        
        # Check upper right diagonal
        i, j = row, col
        while i >= 0 and j < N:
            if board[i][j] == 1:
                return False
            i -= 1
            j += 1
        
        return True
    
    def find_solutions(board, row):
        #recursive function to find solutions
        if row == N:
            # Found solution - make a copy
            solution = []
            for r in range(N):
                row_data = []
                for c in range(N):
                    row_data.append(board[r][c])
                solution.append(row_data)
            solutions.append(solution)
            print(f"Solution {len(solutions)} found!")
            return True
        
        # Try placing queen in each column
        for col in range(N):
            if is_position_safe(board, row, col):
                # Place queen
                board[row][col] = 1
                
                # Recur for next row
                find_solutions(board, row + 1)
                
                # Backtrack
                board[row][col] = 0
    
    print(f"\nSearching solutions for {N}-Queens...")
    find_solutions(board, 0)
    

    print(f"Total solutions found: {len(solutions)}")
    
    # Show first solution
    if solutions:
        print("\nFirst solution (visual representation):")
        print("  " + " ".join([str(i+1) for i in range(N)]))
        for i in range(N):
            print(f"{i+1} " + " ".join(["Q" if x==1 else "." for x in solutions[0][i]]))

# Easy to understand version with comments
def n_queens_simple():
    #Very simple N-Queens implementation
    
    print("|         N-QUEENS SOLVER (SIMPLE)        |")

    
    # Get N from user
    n = int(input("| Enter number of queens: "))
    
    if n <= 0:
        print("Invalid input!")
        return
    
    # Store queen positions: position[row] = column
    position = [-1] * n
    solutions_list = []
    
    def is_valid(row, col):
        #Check if queen can be placed at (row, col)
        for r in range(row):
            c = position[r]
            # Check if same column or diagonal
            if c == col or abs(r - row) == abs(c - col):
                return False
        return True
    
    def place_queens(current_row):
        #Try to place queen at current_row
        if current_row == n:
            # All queens placed successfully
            solutions_list.append(position.copy())
            print(f"\n✓ Solution {len(solutions_list)} found:")
            print(f"  Queen positions (row: column):", end=" ")
            for i in range(n):
                print(f"{i+1}:{position[i]+1}", end="  ")
            return
        
        # Try placing in each column
        for col in range(n):
            if is_valid(current_row, col):
                # Place queen
                position[current_row] = col
                
                # Try next row
                place_queens(current_row + 1)
                
                # Backtrack (remove queen)
                position[current_row] = -1
    
    # Start solving from row 0
    place_queens(0)
    
    # Final summary
   
    print(f"TOTAL SOLUTIONS: {len(solutions_list)}")
    
    if solutions_list:
        # Show first solution as chessboard
        print("\nFIRST SOLUTION (visual):")
        sol = solutions_list[0]
        print("  ", end="")
        for col in range(n):
            print(col+1, end=" ")
        print()
        
        for row in range(n):
            print(row+1, " ", end="")
            for col in range(n):
                if col == sol[row]:
                    print("Q", end=" ")
                else:
                    print(".", end=" ")
            print()

# Run any version you like
if __name__ == "__main__":
    print("Choose version:")
    print("1. Class-based version")
    print("2. Simple version")
    print("3. Alternative version")
    print("4. Very simple version")
    
    choice = input("\nEnter your choice (1-4): ")
    
    if choice == "1":
        solver = NQueensSolver()
        solver.solve()
    elif choice == "2":
        simple_n_ques = simple_n_queens()
    elif choice == "3":
        n_queens_alternative()
    elif choice == "4":
        n_queens_simple()
    else:
        print("Invalid choice!")
```
